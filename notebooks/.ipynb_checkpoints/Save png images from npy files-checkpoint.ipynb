# Convert dataset's npy files into png images

---

```python
import os
import glob

import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

---

```python
def get_temperature_channel(temperature_2d_vector):
    """
    Returns a channel with values between [0, 255] of the same shape as the input (usually 256x256).
    
    Arguments:
        temperature_2d_vector: (256,256) float vector; temperature ranging from [-50, 50].
    """
    norm_img = (temperature_2d_vector + 50) / 100
    
    norm_img[norm_img < 0] = 0   # clip negative numbers
    norm_img[norm_img > 1] = 1   # clip numbers over 1.0
    
    return (norm_img * 255).astype(np.uint8)
```

```python
def get_temperature_2d_vector(temperature_red_channel):
    """
    Returns a (256,256) float vector of temperature ranging from [-50, 50].
    
    Arguments:
        temperature_red_channel: a channel with values between [0, 255] 
                                 extracted from the red channel of a PIL Image.
    """
    return temperature_red_channel / 256 * 100 - 50
```

---

```python
def get_rgb_image_from_temperature(temperature_2d_vector):
    """
    Returns a PIL Image with green and blue channel all 0 and red channel temperature
    
    Arguments:
        temperature_2d_vector: (256,256) float vector; temperature ranging from [-50, 50].
    """
    rgb_channels = np.zeros((256,256,3)).astype(np.uint8)

    t_channel = get_temperature_channel(temperature_2d_vector)
    rgb_channels[:,:,0] = t_channel
    
    return Image.fromarray(rgb_channels)
```

```python
def get_temperature_from_rgb_image(pil_image):
    """
    Returns a 2d vector (256,256) of temperature value [-50, 50] extracted from the red channel of given image
    
    Arguments:
        pil_image: the PIL Image to extract the red channel from
    """
    img_channels = np.array(pil_image)
    return get_temperature_2d_vector(img_channels[:,:,0])
```

```python
def show_images(img1, img2):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,10))
    ax1.imshow(img1)
    #ax1.set_title('Temperature')

    ax2.imshow(img2)
    #ax2.set_title('Red channel Temperature')

    plt.show()
```

## Export all temperatures to rgb images

```python
def export_png_images(source_npy_filenames, dest_foldername, source_res, dates):
    """
    Parse every sample from the specified .npy files and save a png image of the temperature.
    Each png file is named by its date.
    
    Arguments
        source_npy_filenames : array of string, list of .npy files.
        dest_foldername : string, the name of the destination folder where the png images will be saved.
        source_res : 'lowres' or 'hires'
        dates : date in chronological order of the given 'source_npy_filenames'
    """
    # create the destination folder
    if not os.path.exists(dest_foldername):
        os.makedirs(dest_foldername)
    
    temperature_index = 0 if source_res == 'hires' else 5
    date_index = 0
    
    for npy_filename in source_npy_filenames:
        samples = np.load(npy_filename)
        nb_samples = samples.shape[0]
        
        for i in range(nb_samples):
            # create a PIL Image from this sample
            temperature_2d_vector = samples[i,temperature_index,:,:]
            rgb_image = get_rgb_image_from_temperature(temperature_2d_vector)
            
            # save the PIL Image in dest folder using corresponding date as filename
            png_pathname = dest_foldername + '/' + dates[date_index].decode('UTF-8') + '.png'
            rgb_image.save(png_pathname, "PNG")
            
            date_index += 1
            
    return date_index
```
